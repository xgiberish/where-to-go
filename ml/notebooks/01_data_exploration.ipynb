# 01 — Data Exploration
Load `data/raw/destinations.csv`, inspect distributions, identify missing values, and understand the label balance.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/destinations.csv', comment='#')
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
# Label distribution
df['travel_style'].value_counts().plot(kind='bar')
plt.title('Travel Style Distribution')
plt.tight_layout()
plt.show()
```
